```python
df["Lucro Venda"] = df["Valor Venda"] - df["Custo"]
```

```python
df["Dias para Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = True).plot.barh(title="Total de Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro Venda"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.ylabel("Receita")
plt.xticks(rotation='horizontal');
```

```python
df_2009 =  df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.sample(5)
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro Venda"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.ylabel("Lucro Venda");
```

```python
df_2009.groupby("Marca")["Lucro Venda"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro Venda");
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby("Classe")["Lucro Venda"].sum().plot.bar(title="Lucro x Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro Venda");
plt.xticks(rotation='horizontal');
```
